**Missing Value**

```python
# missing values in companies df
companies.isnull().sum()
```

```python
# missing values in rounds df
rounds.isnull().sum()
```

Since there are no misisng values in the permalink or company_permalink columns, let's merge the two and then work on the master dataframe.

```python
# merging the two dataframes
master = pd.merge(companies, rounds, how="inner", left_on="permalink", right_on="company_permalink")
master.head()
```

```python
# print column names
master.columns
```

The columns company_permalink and permalink are the same, let's remove one of them

```python
# removing redundant columns
master =  master.drop(['company_permalink'], axis=1) 
```

```python
# look at columns after dropping
master.columns
```

```python
# column-wise missing values 
master.isnull().sum()
```

```python
# percentage of missing values (column-wise)
round(100*(master.isnull().sum()/len(master.index)), 2)
```

The column funding_round_code is useless (with about 73% missing values). Also, for the business objectives given, the columns homepage_url, founded_at, state_code, region and city need not be used.

```python
# dropping columns 
master = master.drop(columns=['funding_round_code', 'homepage_url', 'founded_at', 'state_code', 'region', 'city'], axis=1)
master.head()
```

```python
# percentage of missing values (column-wise)
round(100*(master.isnull().sum()/len(master.index)), 2)
```

The column raised_amount_usd is an important column, since that is the number we want to analyse (compare, means, sum etc.). That needs to be carefully treated.

Also, the column country_code will be used for country-wise analysis, and category_list will be used to merge the dataframe with the main categories.

```python
# summary stats of raised_amount_usd
master['raised_amount_usd'].describe()
```

The mean is somewhere around USD 10 million, while the median is only about USD 1m. The min and max values has huge difference.

In general, since there is a huge spread in the funding amounts, it will be inappropriate to impute it with a metric such as median or mean. Also, since we have quite a large number of observations, we can drop the rows

```python
# removing NaNs in raised_amount_usd
master=master.dropna( how='any',subset=['raised_amount_usd'])
# or, master = master[~np.isnan(master['raised_amount_usd'])]
```

```python
# percentage of missing values (column-wise)
round(100*(master.isnull().sum()/len(master.index)), 2)
```

Let's now look at the column country_code. To see the distribution of the values for categorical variables, it is best to convert them into type 'category'.

```python
country_codes = master['country_code'].astype('category')

# displaying frequencies of each category
country_codes.value_counts()
```

```python
# viewing fractions of counts of country_codes
100*(master['country_code'].value_counts()/len(master.index))
```

we can either delete the rows having country_code missing (about 6% rows), or we can impute them by USA. Since the number 6 is quite small, and we have a decent amount of data, it may be better to just remove the rows.

```python
# removing rows with missing country_codes
master=master[~pd.isnull(master['country_code'])]

#or, master.dropna( how='any',subset=['country_code'])
```

```python
# look at missing values
round(100*(master.isnull().sum()/len(master.index)), 2)
```

```python
# removing rows with missing category_list values
master = master[~pd.isnull(master['category_list'])]
```

```python
# look at missing values
round(100*(master.isnull().sum()/len(master.index)), 2)
```

**Funding Type Analysis**

We need to impose the constraint that the investment amount should be between 5 and 15 million USD. We will choose the funding type such that the average investment amount falls in this range. Also they want to invest in 4 specified funding type

```python
# first, let's filter the df so it only contains the four specified funding types
sparkfund_df = master[(master.funding_round_type == "venture") | 
        (master.funding_round_type == "angel") | 
        (master.funding_round_type == "seed") | 
        (master.funding_round_type == "private_equity") ]
```

```python
# distribution of raised_amount_usd
sns.boxplot(y=sparkfund_df['raised_amount_usd'])
plt.yscale('log')
plt.show()
```

```python
# summary metrics
sparkfund_df['raised_amount_usd'].describe()
```

There's a significant difference between the mean and the median - USD 9.5m and USD 2m. Let's also compare the summary stats across the four categories.

```python
# comparing summary stats across four categories
sns.boxplot(x='funding_round_type', y='raised_amount_usd', data=sparkfund_df)
plt.yscale('log')
plt.show()
```

```python
# compare the mean and median values across categories
sparkfund_df.pivot_table(values = 'raised_amount_usd', index = 'funding_round_type', aggfunc=[np.median, np.mean])
```

There's a large difference between the mean and the median values for all four types. For type venture, for e.g. the median is about 20m while the mean is about 70m.

Thus, the choice of the summary statistic will drastically affect the decision (of the investment type). Let's choose median, since there are quite a few extreme values pulling the mean up towards them - but they are not the most 'representative' values.

```python
# compare the median investment amount across the types
sparkfund_df.groupby('funding_round_type')['raised_amount_usd'].median().sort_values(ascending=False)
```

The median investment amount for type 'private_equity' is approx. USD 20m, which is beyond Spark Funds' range of 5-15m. The median of 'venture' type is about USD 5m, which is suitable for them. The average amounts of angel and seed types are lower than their range.

Thus, 'venture' type investment will be most suited to them.

**Country Analysis**

```python
# filter the df for venture type investments
sparkfund_df = sparkfund_df[sparkfund_df.funding_round_type=="venture"]

# group by country codes and compare the total funding amounts
country_wise_total =sparkfund_df.groupby('country_code')['raised_amount_usd'].sum().sort_values(ascending=False)
print(country_wise_total)
```

```python
# top 9 countries
top_9_countries = country_wise_total[:9]
top_9_countries
```

Among the top 9 countries, USA, GBR and IND are the top three English speaking countries. Let's filter the dataframe so it contains only the top 3 countries.

```python
# filtering for the top three countries
sparkfund_df = sparkfund_df[(sparkfund_df.country_code=='USA') | (sparkfund_df.country_code=='GBR') | (sparkfund_df.country_code=='IND')]
sparkfund_df.head()
```

```python
sparkfund_df.info()
```

```python
# boxplot to see distributions of funding amount across countries
plt.figure(figsize=(10, 10))
sns.boxplot(x='country_code', y='raised_amount_usd', data=sparkfund_df)
plt.yscale('log')
plt.show()
```

**Sector Analysis**

```python
#extract main category
sparkfund_df['main_category']=sparkfund_df['category_list'].str.split('|').str[0] 
# or, sparkfund_df['category_list'].apply(lambda x: x.split("|")[0])
```

```python
# read mapping file
mapping = pd.read_csv("mapping.csv", sep=",")
mapping.head()
```

```python
# missing values in mapping file
mapping.isnull().sum()
```

```python
# remove the row with missing values
mapping = mapping[~pd.isnull(mapping['category_list'])]
mapping.isnull().sum()
```

```python
# converting common columns to lowercase
mapping['category_list'] = mapping['category_list'].str.lower()
sparkfund_df['main_category'] = sparkfund_df['main_category'].str.lower()
```

```python
# values in main_category column in df which are not in the category_list column in mapping file
sparkfund_df[~sparkfund_df['main_category'].isin(mapping['category_list'])]
```

```python
# values in the category_list column which are not in main_category column 
mapping[~mapping['category_list'].isin(sparkfund_df['main_category'])]
```

There are sectors named alter0tive medicine, a0lytics, waste ma0gement, veteri0ry, etc. This is not a random quality issue, but rather a pattern. In some strings, the 'na' has been replaced by '0'. This is weird - maybe someone was trying to replace the 'NA' values with '0', and ended up doing this.

```python
# replacing '0' with 'na'
mapping['category_list'] = mapping['category_list'].apply(lambda x: x.replace('0', 'na'))
print(mapping['category_list'])
```

```python
#unpivot using melt
mapping = pd.melt(mapping, id_vars=['category_list'])
```

```python
mapping.head()
```

```python
#filter where value is 1
mapping=mapping[mapping.value==1]
```

```python
mapping.head()
```

```python
#drop value column
mapping.drop('value', axis = 1, inplace = True)
```

```python
mapping.head()
```

```python
#rename columns for ease of merging
mapping.rename(columns = {'category_list':'main_category', 'variable':'main_sector'}, inplace = True)
```

```python
mapping.head()
```

```python
master_frame=pd.merge(sparkfund_df, mapping, how = 'left', on = 'main_category')
```

```python
# summarising the sector-wise number and sum of venture investments across three countries

# first, let's also filter for investment range between 5 and 15m
df = master_frame[(master_frame['raised_amount_usd'] >= 5000000) & (master_frame['raised_amount_usd'] <= 15000000)]
```

```python
df.head()
```

```python
# groupby country, sector and compute the count and sum
df.groupby(['country_code', 'main_sector']).raised_amount_usd.agg(['count', 'sum'])
```

```python
# plotting sector-wise count and sum of investments in the three countries
plt.figure(figsize=(16, 14))

sns.barplot(x='main_sector', y='raised_amount_usd', hue='country_code', data=df, estimator=np.sum)
plt.xticks(rotation=20)
plt.title('Total Invested Amount (USD)')
plt.show()

plt.figure(figsize=(16, 14))
sns.countplot(x='main_sector', hue='country_code', data=df)
plt.title('Number of Investments')
plt.xticks(rotation=20)
plt.show()
```

Thus, the top country in terms of the number of investments (and the total amount invested) is the USA. The sectors 'Others', 'Social, Finance, Analytics and Advertising' and 'Cleantech/Semiconductors' are the most heavily invested ones.
